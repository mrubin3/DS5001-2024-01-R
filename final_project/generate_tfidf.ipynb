```python
import pandas as pd
import numpy as np
import configparser
import sys
config = configparser.ConfigParser()
config.read("../../env.ini")
data_home = config['DEFAULT']['data_home']
output_dir = config['DEFAULT']['output_dir']
local_lib = config['DEFAULT']['local_lib']
sys.path.append(local_lib)

OHCO = ['speech_id','para_num', 'sent_num', 'token_num']
bags = dict(
        SENTS = OHCO[:3],
        PARAS = OHCO[:2],
        SPEECHES = OHCO[:1]
)
bag = 'PARAS'
```

```python
DTM = pd.read_csv('DTM.csv').set_index(['speech_id', 'para_num'])
```

```python
DF = DTM.astype('bool').sum()
N = DTM.shape[0]
IDF = np.log2(N / DF)
TF = (DTM.T / DTM.T.sum()).T
TFIDF = TF * IDF
TFIDF
```
