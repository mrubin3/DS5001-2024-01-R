```python
import pandas as pd
import numpy as np
import configparser
import sys
from numpy.linalg import norm
config = configparser.ConfigParser()
config.read("../../env.ini")
data_home = config['DEFAULT']['data_home']
output_dir = config['DEFAULT']['output_dir']
local_lib = config['DEFAULT']['local_lib']
sys.path.append(local_lib)

OHCO = ['speech_id', 'para_num', 'sent_num', 'token_num']
```

```python
VOCAB = pd.read_csv('VOCAB.csv', sep='|').set_index('term_str')
BOW = pd.read_csv('BOW.csv').set_index(['speech_id', 'para_num', 'term_str'])
TFIDF = pd.read_csv('TFIDF.csv').set_index(['speech_id', 'para_num'])
```

```python
short_list = VOCAB[(VOCAB.max_pos_group.isin(['NN', 'VB', 'JJ'])) & (VOCAB.stop == 0)].sort_values('dfidf', ascending=False).head(1500)
TFIDF = TFIDF[short_list.index]
L2 = TFIDF.apply(lambda x: x / norm(x), 1)
L2
```

```python
L2.to_csv('TFIDF_L2.csv')
```
